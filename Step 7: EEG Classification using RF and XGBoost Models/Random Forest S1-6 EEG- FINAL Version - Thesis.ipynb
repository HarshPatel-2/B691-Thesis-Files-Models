# Import Required Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
import sklearn
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, auc, roc_curve, accuracy_score
from sklearn.preprocessing import RobustScaler
import time
import pickle
import os

import warnings
warnings.filterwarnings('ignore')
```

# Load EEG Dataset and Encode Activity Labels

```python
df = pd.read_csv("EEG_Features_Only_S1-6_Final.csv")

LE = LabelEncoder()
df['Activity_Label_OpenBCI'] = LE.fit_transform(df['Activity_Label_OpenBCI'])
```

# Display EEG DataFrame

```python
df
```

# Select Top 101 Important EEG Features and Define Target

```python
# The top 101 most important features
important_features_expanded = [
    'Ch11_Beta', 'Ch9_Beta', 'Ch10_Beta', 'Ch7_Beta', 'Ch16_Beta', 'Ch11_RMS',
    'Ch14_Beta', 'Ch13_Beta', 'Ch11_Std', 'Ch9_Std', 'Ch15_Std', 'Ch10_Std',
    'Ch7_RMS', 'Ch10_Theta', 'Ch15_Beta', 'Ch13_RMS', 'Ch10_Delta', 'Ch11_Delta',
    'Ch16_RMS', 'Ch7_Delta', 'Ch15_RMS', 'Ch10_RMS', 'Ch10_Alpha', 'Ch11_Gamma',
    'Ch9_RMS', 'Ch13_Std', 'Ch11_Alpha', 'Ch14_RMS', 'Ch11_Theta', 'Ch7_Theta',
    'Ch12_Beta', 'Ch6_RMS', 'Ch14_Delta', 'Ch7_Std', 'Ch5_Std', 'Ch6_Std',
    'Ch3_Beta', 'Ch16_Std', 'Ch3_RMS', 'Ch2_Beta', 'Ch8_Std', 'Ch13_Theta',
    'Ch5_RMS', 'Ch13_Gamma', 'Ch15_Delta', 'Ch13_Alpha', 'Ch8_Beta', 'Ch14_Std',
    'Ch7_Alpha', 'Ch6_Beta', 'Ch4_Std', 'Ch10_Gamma', 'Ch8_RMS', 'Ch3_Std',
    'Ch4_RMS', 'Ch13_Delta', 'Ch9_Delta', 'Ch9_Theta', 'Ch5_Delta', 'Ch16_Alpha',
    'Ch6_Delta', 'Ch14_Theta', 'Ch12_Std', 'Ch4_Alpha', 'Ch16_Theta', 'Ch16_Gamma',
    'Ch9_Alpha', 'Ch7_Gamma', 'Ch9_Gamma', 'Ch15_Alpha', 'Ch8_Alpha', 'Ch5_Beta',
    'Ch2_Alpha', 'Ch6_Gamma', 'Ch2_RMS', 'Ch14_Alpha', 'Ch12_Alpha', 'Ch4_Beta',
    'Ch15_Theta', 'Ch12_RMS', 'Ch4_Gamma', 'Ch12_Delta', 'Ch16_Delta', 'Ch12_Gamma',
    'Ch3_Gamma', 'Ch8_Delta', 'Ch6_Alpha', 'Ch3_Delta', 'Ch5_Gamma', 'Ch5_Theta',
    'Ch5_Alpha', 'Ch12_Theta', 'Ch2_Std', 'Ch15_Gamma', 'Ch2_Gamma', 'Ch8_Theta',
    'Ch4_Delta', 'Ch14_Gamma', 'Ch3_Alpha', 'Ch8_Gamma', 'Ch6_Theta'
]

PredictorCol = important_features_expanded
TargetCol = 'Activity_Label_OpenBCI'

X = df[PredictorCol].values
y = df[TargetCol].values
```

# Split EEG Data into Training and Testing Sets (Stratified)

```python
X_train_expanded, X_test_expanded, y_train_expanded, y_test_expanded = train_test_split(
    X, y, test_size=0.30, random_state=15, stratify=y
)

print(f"X_train shape: {X_train_expanded.shape}")
print(f"X_test shape: {X_test_expanded.shape}")
```

# Train and Evaluate Random Forest on Expanded EEG Features

```python
from sklearn import metrics

# Initialize the RandomForestClassifier with the best hyperparameters found
clf_expanded = RandomForestClassifier(
    n_estimators=200,          
    criterion='gini',
    class_weight='balanced',   
    max_depth=18,              
    min_samples_split=2,       
    min_samples_leaf=1        
)

# ===================== Model Training and Timing =====================

# Measure the time to train the model
start_train_time = time.time()

# Train the model
RF_expanded = clf_expanded.fit(X_train_expanded, y_train_expanded)

# Measure training time
train_time = time.time() - start_train_time

# Display training time
print("===== Training and Inference Times =====")
print(f"Training Time: {train_time:.4f} seconds")

# ===================== Prediction and Inference Timing =====================

# Measure the inference time (time to predict on test set)
start_infer_time = time.time()
RF_expanded.predict(X_test_expanded)
infer_time = time.time() - start_infer_time

# Display inference time
print(f"Inference Time (for test set): {infer_time:.4f} seconds")

# ===================== Accuracy Calculation =====================

# Calculate Train and Test Accuracy
train_preds_expanded = RF_expanded.predict(X_train_expanded)
test_preds_expanded = RF_expanded.predict(X_test_expanded)

train_acc_expanded = accuracy_score(y_train_expanded, train_preds_expanded)
test_acc_expanded = accuracy_score(y_test_expanded, test_preds_expanded)

# ===================== Stratified K-Fold Cross-Validation =====================

# Create StratifiedKFold object
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Cross-validation using StratifiedKFold
stratified_scores = cross_val_score(RF_expanded, X_train_expanded, y_train_expanded, cv=skf)

# Calculate the mean and standard deviation of Stratified K-Fold results
mean_stratified_score = stratified_scores.mean()
std_stratified_score = stratified_scores.std()

# =========================== Combined Results ===========================

print("\n===== Stratified Cross-Validation Results =====")
print(f"Stratified CV Accuracy (mean ± std): {mean_stratified_score:.4f} ± {std_stratified_score:.4f}")
print(f"Training Set Accuracy: {train_acc_expanded:.4f}")
print(f"Test Set Accuracy: {test_acc_expanded:.4f}")

# ===================== Classification Report =====================

# Get original class names from the encoder
class_names = LE.classes_

print("\n===== Classification Report =====")
print(metrics.classification_report(y_test_expanded, test_preds_expanded, target_names=class_names))

# ===================== Confusion Matrix =====================

# Generate confusion matrix
conf_matrix_expanded = metrics.confusion_matrix(y_test_expanded, test_preds_expanded)

# Convert confusion matrix to DataFrame for better readability
df_conf_matrix_expanded = pd.DataFrame(conf_matrix_expanded, index=class_names, columns=class_names)

print("\n===== Confusion Matrix on Test Set =====")
print(df_conf_matrix_expanded)

# ======================= Plot Confusion Matrix ========================

# Plot the confusion matrix using seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_expanded, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title('Random Forest (Expanded Features) - Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

# Save Trained EEG Random Forest Model

```python
import pickle

with open("RF_S1_6_EEG_expanded.pkl", "wb") as file:
    pickle.dump(RF_expanded, file)

print("Random Forest model for S1-6 EEG data has been successfully saved to 'RF_S1_6_EEG_expanded.pkl'")
```
