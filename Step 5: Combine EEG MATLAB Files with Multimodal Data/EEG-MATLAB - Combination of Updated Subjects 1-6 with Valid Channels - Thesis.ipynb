# Import Libraries

```python
import pandas as pd
```

# Preview Individual EEG CSV Files

```python
files = [
    "EEG_Features_Only_S1-ValidChannels_Updated.csv",
    "EEG_Features_Only_S2-ValidChannels_Updated.csv",
    "EEG_Features_Only_S3-ValidChannels_Updated.csv",
    "EEG_Features_Only_S4-ValidChannels_Updated.csv",
    "EEG_Features_Only_S5-ValidChannels_Updated.csv",
    "EEG_Features_Only_S6-ValidChannels_Updated.csv"
]

for file in files:
    df = pd.read_csv(file)
    rows, cols = df.shape
    print(f"\n{file} -> Rows: {rows}, Columns: {cols}")
    print(f"Head of {file}:")
    print(df.head())
```

# Combine and Save EEG CSV Files

```python
files = [
    "EEG_Features_Only_S1-ValidChannels_Updated.csv",
    "EEG_Features_Only_S2-ValidChannels_Updated.csv",
    "EEG_Features_Only_S3-ValidChannels_Updated.csv",
    "EEG_Features_Only_S4-ValidChannels_Updated.csv",
    "EEG_Features_Only_S5-ValidChannels_Updated.csv",
    "EEG_Features_Only_S6-ValidChannels_Updated.csv"
]

dataframes = [pd.read_csv(file) for file in files]
combined_df = pd.concat(dataframes, ignore_index=True)

rows, cols = combined_df.shape
print(f"Combined Dataset -> Rows: {rows}, Columns: {cols}")

print("\nHead of Combined Dataset:")
print(combined_df.head())

combined_df.to_csv("EEG_Features_Only_S1-6_Final.csv", index=False)
print("\nSaved combined dataset as 'EEG_Features_Only_S1-6_Final.csv'")
```
